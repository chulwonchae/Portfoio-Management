# Estimating Vol
## 1. Rolling Vol

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
sns.set()

etf = yf.Ticker('SPY')
data = etf.history(start='2000-01-01', actions=False)
data.drop(['Open', 'High', 'Low', 'Volume'], inplace=True, axis=1)

rets = data.pct_change().fillna(0)
```

### 20 days rolling vol


```python
param = 20
vols = rets.rolling(param).std()*np.sqrt(252)
vols.plot()
```

### 60 days rolling vol

```python
param = 60
vols_60 = rets.rolling(param).std() * np.sqrt(252)
vols_60.plot()
```

## 2. EWMA(Exponential Weighted Moving Average) Vol

```python
alpha = 0.2
vols = rets.ewm(alpha=alpha).std() * np.sqrt(252)
vols.plot()
```

```python
alpha = 0.9
vols = rets.ewm(alpha=alpha).std() * np.sqrt(252)
vols.plot()
```
