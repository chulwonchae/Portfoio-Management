The history() method accepts several parameters that allow you to customize the data query process. Below are the parameters of the history() method along with their descriptions:

- period: Sets the period for which the data will be downloaded. The default is one month (1mo). Options include (1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max).
- interval: Sets the frequency of the data. If you choose an intraday interval, data is available for up to 60 days. The default is one day (1d). Options include (1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo).
- start: Sets the start date of the data if the period parameter is not used. You can use a string in 'YYYY-MM-DD' format or a datetime object.
- end: Sets the end date of the data if the period parameter is not used. You can use a string in 'YYYY-MM-DD' format or a datetime object.
- prepost: Indicates whether to include pre-market and after-hours trading data in the results. The default is False.
- auto_adjust: Indicates whether to automatically use adjusted prices. The default is True.
- actions: For stocks, this parameter sets whether to download corporate actions such as dividends or stock splits. The default is True.

### Other Data

`yfinance`는 가격 데이터 뿐만 아니라 재무 데이터, 대주주 보유 현황, 애널리스트 의견, 뉴스 데이터 또한 제공합니다. 아래의 리스트는 `yfinance`를 통해 얻을 수 있는 주요 데이터를 가져올 수 있는 속성값들입니다.

```python
import yfinance as yf

# Fetch the data for Apple Inc. (AAPL)
aapl = yf.Ticker('AAPL')

# Get the quarterly financial data
quarterly_financials = aapl.quarterly_financials
annual_financials = aapl.financials

# Print the quarterly financial data
print("Quarterly Financial Data:")
print(quarterly_financials)

# Print the annual financial data
print("\nAnnual Financial Data:")
print(annual_financials)
```

```python
# Extract net income (earnings) and revenue data
net_income = quarterly_financials.loc['Net Income']
revenue = quarterly_financials.loc['Total Revenue']

print("Net Income (quarterly):")
print(net_income)

print("\nTotal Revenue (quarterly):")
print(revenue)
```
